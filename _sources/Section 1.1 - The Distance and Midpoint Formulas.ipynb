#### 1. Find the distance $d(P_1, P_2)$ between the points $P_1$ and $P_2$: $P_1 = (a,a); P_2 = (0,0)$

Solution:

$ 
\begin{align}
d(P_1, P_2) &= \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2} \\
&= \sqrt{(a - 0)^2 + (a - 0)^2}  \\
&= \sqrt{2a^2}  \\
&= a\sqrt 2
\end{align}
$

#### 2. Plot each point and form the triangle $ABC$. Verify that the triangle is a right triangle. Find its area.

$
A = (-2,5) \\
B = (12,3) \\
C = (10,-11)
$

Solution:

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots()

x_coordinates = [-2, 12, 10, -2]
y_coordinates = [5, 3, -11, 5]

plt.scatter(x_coordinates, y_coordinates)
plt.plot(x_coordinates, y_coordinates)
ax.annotate("A", xy = (-2,2), size = 20)
plt.annotate("B", xy = (10,1), size = 20)
plt.annotate("c", xy = (11,-11), size = 20)
```

$ 
\begin{align}
d(AB) &= \sqrt{(-2 - 12)^2 + (5 - 3)^2}  \\
&= \sqrt{196 + 4}  \\
&= 10\sqrt2 
\end{align}
$


$ 
\begin{align}
d(BC) &= \sqrt{(12 - 10)^2 + (3 - (-11))^2}  \\
&= \sqrt{4 + 196}  \\
&= 10\sqrt2 
\end{align}
$


$ 
\begin{align}
d(CA) &= \sqrt{(10 - (-2))^2 + (-11 - 5)^2}  \\
&= \sqrt{144 + 256}  \\
&= 20
\end{align}
$

$
\begin{align}
CA^2 &= AB^2 + BC^2  \\
400 &= 200 + 200
\end{align}
$

Hence, $\Delta ABC$ is a right triangle.

Area of $\Delta ABC$ = $ \dfrac{1}{2} \cdot AB \cdot BC$ = $ \dfrac{1}{2} \cdot (10\sqrt2) \cdot (10\sqrt2)= 100 $

#### 3. The median of a triangle are the line segments from each vertex to the midpoint of the opposite side. Find the lengths of the medians of the triangle with vertices at $A = (0,0), B = (6,0), \text{ and } C(4,4)$. 

Solution:

$
E = \bigg ( \dfrac{0 + 6}{2}, \dfrac{0 + 0}{2} \bigg ) = (3,0) \\
F = \bigg ( \dfrac{4 + 6}{2}, \dfrac{0 + 4}{2} \bigg ) = (5,2) \\ 
G = \bigg ( \dfrac{0 + 4}{2}, \dfrac{0 + 4}{2} \bigg ) = (2,2) \\ 
$

```python
fig, ax = plt.subplots()

x_coordinates = [0, 6, 4, 0]
y_coordinates = [0, 0, 4, 0]

plt.scatter(x_coordinates, y_coordinates)
plt.scatter([(0+6)/2, (6+4)/2, (4+0)/2], [(0+0)/2, (0+4)/2, (4+0)/2])
plt.plot(x_coordinates, y_coordinates)
plt.annotate("A", xy = (0,.5), size = 20)
plt.annotate("C", xy = (3.75,3.5), size = 20)
plt.annotate("B", xy = (5.5,.25), size = 20)
plt.annotate("E", xy = (3,.5), size = 20)
plt.annotate("F", xy = (5.25,2), size = 20)
plt.annotate("G", xy = (1.75,2.25), size = 20)
```

$
\text{CE = the median of AB} = \sqrt{(4-3)^2 + (4-0)^2 } = \sqrt{17} \\
\text{AF = the median of BC} = \sqrt{(5-0)^2 + (2-0)^2 } = \sqrt{29} \\
\text{BG = the median of AC} = \sqrt{(6-2)^2 + (0-2)^2 } = \sqrt{20} = 2\sqrt5 \\
$

#### 4. If two vertices of an equilateral triangle are $(0,4)$ and $(0,0)$, find the third vertex. How many of these triangles are possible?

Solution:

```python
fig, ax = plt.subplots()

x_coordinates = [0, 0, .6, 0]
y_coordinates = [4, 0, 2, 4]

plt.scatter(x_coordinates, y_coordinates)
plt.plot(x_coordinates, y_coordinates)
ax.annotate("A", xy = (0.05,.5), size = 20)
plt.annotate("B", xy = (.05, 3.25), size = 20)
plt.annotate("(x,y)", xy = (.45, 2), size = 20)
plt.annotate("C", xy = (.55, 1.5), size = 20)
```

For the equilateral triangle $\Delta ABC$ we can write:

$
AB^2 = BC^2 = AC^2  \\
4^2 = (x-0)^2 + (y-4)^2 = (x-0)^2 + (y-0)^2
$

$
x^2 + (y-4)^2 = x^2 + y^2  \\
y^2 - 8y + 16 = y^2  \\
8y = 16 \\
y = 2
$


$
x^2 + 4 = 16  \\
x = \sqrt 12  \\
x = \pm 2\sqrt3
$

Therefore, there are two triangles possible with the third vertex at $(2\sqrt3,2)$ or at $(-2\sqrt3,2)$.

#### 5. A Ford Focus and a Freightliner truck leave an intersection at the same time. The focus heads east at an average speed of 30 miles per hr, while the truck heads south at an average speed of 40 miles per hr. Find an expression for their distance apart $d$ (in miles) at the end of $t$ hours.  

Solution:

```python
fig, ax = plt.subplots()

x_coordinates = [0,  4,   0,  0]
y_coordinates = [0,  0,  -4,  0]

plt.scatter(x_coordinates, y_coordinates)
plt.plot(x_coordinates, y_coordinates)
plt.annotate("30t", xy = (2,-.5), size = 20)
plt.annotate("40t", xy = (0,-2), size = 20)
plt.annotate("d", xy = (2,-2.5), size = 20)
```

After $t$ hours, the car moves $30t$ miles to the east.
After $t$ hours, the truck moves $40t$ miles to the south.

Since the vehicle move at right angle with each other, we can write:

$
d^2 = (30t)^2 + (40t)^2  \\
d^2 = 2500t^2 \\
d = 50t
$
